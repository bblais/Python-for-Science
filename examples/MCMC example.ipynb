## Trying to fit a poisson process

```python
%matplotlib inline
from science import *
from scipy.stats import poisson
```

```python
x=np.random.poisson(lam=10,size=1000)
```

```python
def lnprior(mu):
    if 0<=mu<=100:
        return 0.0
    return -np.inf
```

```python
def lnlike(data,mu):
    return log(mu)*sum(data)-mu*len(data)
```

```python
model=MCMCModel2(x,lnprior,lnlike,
                mu=Uniform(0,100))
```

```python
model.run_mcmc(500)
model.plot_chains()
```

```python
model.plot_distributions()
```
